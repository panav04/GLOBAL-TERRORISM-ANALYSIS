## 3. ***Data Wrangling***

### Data Wrangling Code

````python
try:
    if 'iyear' in df.columns:
        columns_to_keep = [
            'iyear', 'imonth', 'iday', 'country_txt', 'region_txt',
            'provstate', 'city', 'latitude', 'longitude',
            'attacktype1_txt', 'targtype1_txt', 'weaptype1_txt',
            'gname', 'nkill', 'nwound'
        ]

        df = df[columns_to_keep].copy()

        rename_mapping = {
            'iyear': 'Year', 'imonth': 'Month', 'iday': 'Day',
            'country_txt': 'Country', 'region_txt': 'Region',
            'provstate': 'State', 'city': 'City',
            'latitude': 'Latitude', 'longitude': 'Longitude',
            'attacktype1_txt': 'Attack_Type', 'targtype1_txt': 'Target_Type',
            'weaptype1_txt': 'Weapon_Type', 'gname': 'Group_Name',
            'nkill': 'Killed', 'nwound': 'Wounded'
        }
        df.rename(columns=rename_mapping, inplace=True)

        df['Killed'] = df['Killed'].fillna(0)
        df['Wounded'] = df['Wounded'].fillna(0)
        df['State'] = df['State'].fillna('Unknown')
        df['City'] = df['City'].fillna('Unknown')

        df['Total_Casualties'] = df['Killed'] + df['Wounded']

        print("Data Wrangling successful! Dataset is now analysis-ready.")
        print(f"New dataset shape: {df.shape[0]} rows and {df.shape[1]} columns.")
        display(df.head(3))

    else:
        print("Data is already wrangled and ready for analysis! No changes made.")
        display(df.head(3))

except Exception as e:
    print(f"An error occurred during Data Wrangling: {e}")
````

### What all manipulations have you done and insights you found?

# Manipulations Done:

1. Column Filtering (Dimensionality Reduction): I reduced the dataset from 135 columns down to the 15 core columns essential for our business objective. I utilized .copy() to create an independent dataframe, preventing Pandas memory allocation errors (SettingWithCopyWarning).

2. Column Renaming: I renamed obscure database headers (e.g., iyear, nkill, country_txt) into standard English terms (e.g., Year, Killed, Country) to ensure our future visualizations are stakeholder-friendly and easy to read.

3. Missing Value Imputation: I replaced missing (NaN) values in the numerical Killed and Wounded columns with 0 to ensure mathematical operations do not fail. For text-based columns like State and City, I replaced missing values with "Unknown".

4. Feature Engineering: I created a new metric column named Total_Casualties by adding the Killed and Wounded columns together. This provides a single, unified metric to evaluate the severity of an attack.

5. Idempotency (Production-Grade Logic): I wrapped the entire wrangling block in a conditional statement (if 'iyear' in df.columns:) so the cell can be executed multiple times without throwing KeyError exceptions.

# Insights Found During Wrangling:

1. (Data Integrity): I discovered that attempting to perform arithmetic on columns with NaN values results in broken data. By proactively imputing these with 0, I secured the mathematical integrity of our Total_Casualties column.

2. (Computational Efficiency): The original dataset was bloated with 100+ empty or irrelevant columns. By aggressively filtering them out, the dataset is now significantly lighter in memory. This means our upcoming Matplotlib and Seaborn charts will render much faster.

3. (Execution Flow): I learned that Jupyter Notebooks can crash if a renaming cell is run twice. Implementing conditional execution makes the pipeline robust and deployment-ready.

#### Chart - 3

````python
# BAR PLOT - Most Frequent Attack Types

try:
    attack_counts = df['Attack_Type'].value_counts()

    total_attacks = attack_counts.sum()

    plt.figure(figsize=(14, 7))

    ax = sns.barplot(x=attack_counts.values, y=attack_counts.index, palette='crest')

    plt.title('Distribution of Terrorist Attack Types', fontsize=16, fontweight='bold')
    plt.xlabel('Number of Attacks', fontsize=12)
    plt.ylabel('Attack Type', fontsize=12)

    for i, p in enumerate(ax.patches):
        percentage = f'{100 * p.get_width() / total_attacks:.1f}%'
        ax.annotate(percentage,
                    (p.get_width(), p.get_y() + p.get_height() / 2.),
                    ha='left', va='center',
                    xytext=(5, 0),
                    textcoords='offset points',
                    fontsize=11, fontweight='bold', color='black')

    sns.despine()
    plt.show()

except Exception as e:
    print(f"An error occurred while generating Chart 3: {e}")
````
